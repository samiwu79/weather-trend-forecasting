# 02 Exploratory Data Analysis (EDA) for Weather Trends 

This notebook performs basic EDA on the cleaned daily dataset produced by 01_data_cleaning.
Charts use matplotlib only (one chart per cell, no custom colors).

 1) Load the cleaned data
 2) Inspect structure & missingness
 3) Plot distributions (temperature, precipitation)
 4) Explore monthly time trends (global)
 5) Compare countries (top-N)
 6) Correlation heatmap (numeric features)
 7) Save charts to the `visuals/` folder for your report


#  1) Load cleaned daily data

```python
import pandas as pd
from pathlib import Path

# Try common relative paths so it works whether you run from project root or from notebooks/
candidate_paths = [Path("data/cleaned_weather.csv"),
                   Path("../data/cleaned_weather.csv")]
CLEAN = None
for p in candidate_paths:
    if p.exists():
        CLEAN = p
        break
assert CLEAN is not None, "cleaned_weather.csv not found. Please run 01_data_cleaning first."
print("Using:", CLEAN.resolve())

df = pd.read_csv(CLEAN, parse_dates=["timestamp"])
print(df.shape)
df.head()
```

# 2) Basic structure and statistics

```python
# Inspect data types and a quick statistical summary
df.info()

# Full describe (numeric + non-numeric); this can be long
df.describe(include="all")

# Missingness per column (ratio), sorted
missing_ratio = df.isna().mean().sort_values(ascending=False)
print(missing_ratio.head(20))
```

# 3) Choose key columns for EDA

```python
# Set your actual column names here (based on your dataset)
temp_col = "temperature_celsius"   # temperature column in your data
precip_col = "precip_mm"           # precipitation column in your data (may be missing sometimes)
time_col = "timestamp"             # time column
geo_col = "country"                # geography (no 'city' in your dataset, we use 'country')

# Sanity checks
assert temp_col in df.columns, f"Missing temperature column: {temp_col}"
assert time_col in df.columns, f"Missing time column: {time_col}"
assert geo_col in df.columns, f"Missing geography column: {geo_col}"
has_precip = precip_col in df.columns
print("Columns OK. has_precip =", has_precip)
```

#  4) Prepare an output folder for charts

```python
if str(CLEAN).startswith("../"):
    VIS = Path("../visuals")
else:
    VIS = Path("visuals")
VIS.mkdir(parents=True, exist_ok=True)
print("Saving figures to:", VIS.resolve())
```

# 5) Distributions — Temperature

```python
import matplotlib.pyplot as plt

plt.figure()
df[temp_col].hist(bins=50)
plt.title("Temperature distribution")
plt.xlabel(temp_col)
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(VIS / "hist_temperature.png")
plt.show()
```

#  5b) Distributions — Precipitation

```python
if has_precip:
    plt.figure()
    df[precip_col].hist(bins=50)
    plt.title("Precipitation distribution")
    plt.xlabel(precip_col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig(VIS / "hist_precipitation.png")
    plt.show()
else:
    print("No precipitation column; skipping precip histogram.")
```

# 6) Time trends (monthly global averages)

```python
# Build monthly averages across all countries for clarity
monthly = (
    df.groupby(df[time_col].dt.to_period("M"))
      .mean(numeric_only=True)
      .reset_index()
)
# Convert Period to Timestamp for plotting
monthly[time_col] = monthly[time_col].dt.to_timestamp()
monthly.head()
```

#  6a) Monthly temperature trend (global)

```python
plt.figure()
plt.plot(monthly[time_col], monthly[temp_col])
plt.title("Monthly Average Temperature (Global)")
plt.xlabel("Month")
plt.ylabel(temp_col)
plt.tight_layout()
plt.savefig(VIS / "trend_monthly_temperature.png")
plt.show()
```


#  6b) Monthly precipitation trend (global)

```python
if has_precip and (precip_col in monthly.columns):
    plt.figure()
    plt.plot(monthly[time_col], monthly[precip_col])
    plt.title("Monthly Average Precipitation (Global)")
    plt.xlabel("Month")
    plt.ylabel(precip_col)
    plt.tight_layout()
    plt.savefig(VIS / "trend_monthly_precipitation.png")
    plt.show()
else:
    print("No precipitation for monthly plotting; skipping.")
```

#  7) Country comparisons — Top-N countries

```python
# Pick top 10 countries by row count (data-rich)
top_n = 10
top_countries = df[geo_col].value_counts().index[:top_n]

# Average temperature by country (top-N)
avg_by_country = (
    df[df[geo_col].isin(top_countries)]
      .groupby(geo_col)[temp_col]
      .mean()
      .sort_values()
)
print(avg_by_country.tail(10))

plt.figure()
avg_by_country.plot(kind="barh")
plt.title(f"Average Temperature by Country (Top {len(avg_by_country)})")
plt.xlabel(temp_col)
plt.ylabel(geo_col)
plt.tight_layout()
plt.savefig(VIS / "bar_temperature_by_country_topN.png")
plt.show()

# Precipitation by country (top-N), if present
if has_precip:
    avg_precip_by_country = (
        df[df[geo_col].isin(top_countries)]
          .groupby(geo_col)[precip_col]
          .mean()
          .sort_values()
    )
    plt.figure()
    avg_precip_by_country.plot(kind="barh")
    plt.title(f"Average Precipitation by Country (Top {len(avg_precip_by_country)})")
    plt.xlabel(precip_col)
    plt.ylabel(geo_col)
    plt.tight_layout()
    plt.savefig(VIS / "bar_precip_by_country_topN.png")
    plt.show()
else:
    print("No precipitation column; skipping precipitation-by-country chart.")
```

#  8) Correlation analysis (numeric variables only)

```python
import numpy as np

numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure()
im = plt.imshow(corr.values, aspect="auto")
plt.title("Correlation Heatmap (Numeric Columns)")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.savefig(VIS / "correlation_heatmap.png")
plt.show()
```

# 9) Save a quick text summary (useful for README/report)

```python
summary_lines = []
summary_lines.append(f"Rows: {len(df)}")
summary_lines.append(f"Columns: {len(df.columns)}")
summary_lines.append(f"Time range: {df['timestamp'].min()} → {df['timestamp'].max()}")
summary_lines.append(f"Countries: {df['country'].nunique()}")
summary_lines.append("Top 5 countries: " + ", ".join(df['country'].value_counts().index[:5].tolist()))

summary_path = (VIS / "eda_quick_summary.txt")
with open(summary_path, "w", encoding="utf-8") as f:
    f.write("\n".join(summary_lines))
print("Saved summary to:", summary_path.resolve())
```
