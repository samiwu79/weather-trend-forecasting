# 03 — Basic Forecasting (Temperature) with timestamp

 Will build a simple yet solid baseline for forecasting **daily temperature**:
 1) Load cleaned daily data from 01
 2) Choose a target series (Global or a Country)
 3) Create several baseline models:
    - Naive (last value)
    - Seasonal Naive (last value from same weekday, period=7)  — if enough history
    - Moving Average (window=7)
 4) Fit a small ARIMA (grid over p,d,q in {0,1,2})
 5) Forecast for the test period (last N days), compare with actuals
 6) Evaluate with MAE / RMSE / MAPE
 7) Plot predictions vs. actuals, save figures and metrics

#  1) Imports & paths

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA

# Locate cleaned data whether running from project root or notebooks/
candidate_paths = [Path("data/cleaned_weather.csv"),
                   Path("../data/cleaned_weather.csv"),
                   Path("../../data/cleaned_weather.csv")]
CLEAN = None
for p in candidate_paths:
    if p.exists():
        CLEAN = p
        break
assert CLEAN is not None, "cleaned_weather.csv not found. Please run 01_data_cleaning first."
print("Using cleaned file:", CLEAN.resolve())

# Resolve output folders
if str(CLEAN).startswith("../"):
    VIS = Path("../visuals")
    REP = Path("../report")
else:
    VIS = Path("visuals")
    REP = Path("report")
VIS.mkdir(parents=True, exist_ok=True)
REP.mkdir(parents=True, exist_ok=True)
```


#  2) Load data & choose columns

```python
df = pd.read_csv(CLEAN, parse_dates=["timestamp"])
# Your dataset columns (from your 02 inspection)
time_col = "timestamp"
geo_col  = "country"
temp_col = "temperature_celsius"

for col in [time_col, geo_col, temp_col]:
    assert col in df.columns, f"Missing column: {col}"
print(df.shape, "columns:", df.columns.tolist()[:10], "...")
```

#  3) Choose target series: Global or a Country

- Global: Take the average of daily weather data from all countries and synthesize a "Global average temperature/precipitation" sequence.
- country: Only take the daily data of a certain country (rather than the average of all countries).


```python
# --- OPTION A: GLOBAL average (across all countries) ---
global_daily = (df
                .groupby(time_col, as_index=False)
                .mean(numeric_only=True))

# --- OPTION B: Country-level (pick the most data-rich country) ---
top_country = df[geo_col].value_counts().index[0]
country_daily = (df[df[geo_col] == top_country]
                 .sort_values(time_col))

print("Top country (by count):", top_country)

# ====== >>> CHOOSE ONE SERIES TO MODEL <<< ======

TARGET = "global"   # set to "global" or "country"
if TARGET == "global":
    series_df = global_daily[[time_col, temp_col]].copy()
    title_prefix = "Global"
else:
    series_df = country_daily[[time_col, temp_col]].copy()
    title_prefix = f"Country: {top_country}"
```


#  4) Ensure daily frequency & basic preparation

```python
# - Set timestamp as index
# - Sort by time
# - Set daily frequency (D)
# - Fill small gaps with forward-fill (ffill)
s = series_df.sort_values(time_col).set_index(time_col)[temp_col].asfreq("D")
s = s.ffill()  # forward-fill if there are small missing days
print("Series length:", len(s), "from", s.index.min().date(), "to", s.index.max().date())

# Plot the series
plt.figure()
plt.plot(s.index, s.values)
plt.title(f"{title_prefix} — Daily {temp_col}")
plt.xlabel("Date")
plt.ylabel(temp_col)
plt.tight_layout()
plt.savefig(VIS / f"{title_prefix.lower().replace(':','')}_daily_series.png")
plt.show()
```

#  5) Train / Test split

- `train` the model with train data
- `test` data to verify the predictive ability of the model (whether the prediction for the next 30 days is accurate).

```python
# will use the last N days as test (forecast horizon)
H = 30  # forecast horizon (days)
assert len(s) > H + 30, "Not enough data. Ensure you have >60 days to hold out 30 days for test."

train = s.iloc[:-H]
test  = s.iloc[-H:]

print("Train:", train.index.min().date(), "→", train.index.max().date(), "len:", len(train))
print("Test :", test.index.min().date(),  "→", test.index.max().date(),  "len:", len(test))
```

# 6) Helper metrics

- `MAE` (MEAN ABSOLUTE ERROR):  How many "temperature units" (℃) do the predicted values and the true values differ on average?
- `RMSE`（Root Mean Squared Error): the root of mae
- `MAPE`（Mean Absolute Percentage Error):  Describe the size of the average error as a percentage to facilitate cross-regional/dimensional comparisons.


```python
def mae(y_true, y_pred):
    return float(np.mean(np.abs(np.asarray(y_true) - np.asarray(y_pred))))

def rmse(y_true, y_pred):
    return float(np.sqrt(np.mean((np.asarray(y_true) - np.asarray(y_pred))**2)))

def mape(y_true, y_pred):
    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    eps = 1e-9
    return float(np.mean(np.abs((y_true - y_pred) / (y_true + eps))) * 100.0)
```

# Baseline

- `Baseline 1-Naive Forecast`: The predicted values for all future days = the true values on the last day of the training set.
- `Baseline 2 — Seasonal Naive Forecast`: The predicted value for a certain future day = the true value of the same day a week ago.

```python
#  Baseline 1: Naive (last value)
# Forecast every day in test as the last training value
yhat_naive = np.full(shape=len(test), fill_value=train.iloc[-1])
mae_naive  = mae(test, yhat_naive)
rmse_naive = rmse(test, yhat_naive)
mape_naive = mape(test, yhat_naive)

# Baseline 2: Seasonal Naive (period=7), if enough history
# yhat[t] = train[t-7] (last same weekday)
if len(train) >= 7:
    yhat_snaive = []
    hist = train.copy()
    for i in range(len(test)):
        idx = i - 7
        if len(hist) + idx >= 0:
            yhat_snaive.append(hist.iloc[idx])
        else:
            # Fallback to last value
            yhat_snaive.append(hist.iloc[-1])
    yhat_snaive = np.array(yhat_snaive)
    mae_snaive  = mae(test, yhat_snaive)
    rmse_snaive = rmse(test, yhat_snaive)
    mape_snaive = mape(test, yhat_snaive)
else:
    yhat_snaive = None
    mae_snaive = rmse_snaive = mape_snaive = np.nan

```

#  10) ARIMA (small grid search over p,d,q in {0,1,2})

```python
# Keep it tiny to be fast & beginner-friendly
best_order = None
best_aic = np.inf
best_model = None

for p in [0,1,2]:
    for d in [0,1,2]:
        for q in [0,1,2]:
            try:
                model = ARIMA(train, order=(p,d,q))
                res = model.fit()
                if res.aic < best_aic:
                    best_aic = res.aic
                    best_order = (p,d,q)
                    best_model = res
            except Exception:
                continue

print("Best ARIMA order:", best_order, "AIC:", best_aic)

# Forecast H steps
if best_model is not None:
    arima_forecast = best_model.forecast(steps=H)
    yhat_arima = np.asarray(arima_forecast)
    mae_arima  = mae(test, yhat_arima)
    rmse_arima = rmse(test, yhat_arima)
    mape_arima = mape(test, yhat_arima)
else:
    yhat_arima = None
    mae_arima = rmse_arima = mape_arima = np.nan
```

#  11) Collect metrics

#  12) Plot: Predictions vs Actuals

```python
# Build a dataframe for plotting
plot_df = pd.DataFrame({
    "actual": test.values,
    "naive": yhat_naive,
    "ma7": yhat_ma
}, index=test.index)

if not np.isnan(mae_snaive):
    plot_df["snaive7"] = yhat_snaive
if not np.isnan(mae_arima):
    plot_df["arima"]   = yhat_arima

plt.figure()
plt.plot(plot_df.index, plot_df["actual"], label="Actual")
for col in plot_df.columns:
    if col == "actual": continue
    plt.plot(plot_df.index, plot_df[col], label=col)
plt.title(f"{title_prefix} — Forecast vs Actuals (H={H} days)")
plt.xlabel("Date")
plt.ylabel(temp_col)
plt.legend()
plt.tight_layout()
out_path = VIS / f"{title_prefix.lower().replace(':','')}_forecast_vs_actuals.png"
plt.savefig(out_path)
plt.show()
print("Saved plot to:", out_path.resolve())
```
